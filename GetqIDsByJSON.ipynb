```python
import requests
from bs4 import BeautifulSoup as bs
import types
import re
import pymysql
import time
import sys
import json
def dttounix(dt):
    return int(time.mktime(time.strptime(dt, '%Y-%m-%d %H:%M:%S')))
# Qlist[ [qid,title,tag,owner_id,score,is_answered,create_time,last_active_time],[]... ]
def GetQuestionInfo(s,e,m):
    Qlist=[]
    start=dttounix(s)
    end=dttounix(e)
    page=1
    psize=100 #一頁有100個question object
    minscore=m
    while True: #每次迴圈就是翻一頁
        url="https://api.stackexchange.com/2.2/questions?page="+str(page)+"&pagesize="+str(psize)+"&fromdate="+str(start)+"&todate="+str(end)+"&order=desc&min="+str(minscore)+"&sort=votes&site=stackoverflow"
        text=requests.get(url=url).json()
        for question in text['items']:
            qid=question['question_id']
            title=question['title']
            tags=question['tags']
            if 'user_id' in question['owner']:
                owner_id=question['owner']['user_id']
            else: owner_id=0
            is_answered=question['is_answered']
            create_time=question['creation_date']
            last_active_time=question['last_activity_date']
            Qlist.append([qid,title,tags,owner_id,is_answered,create_time,last_active_time])
        if text["has_more"]: page+=1
        else: break
    return Qlist
QuestionInfo=GetQuestionInfo('2016-03-20 00:00:00','2016-03-25 23:59:59',20)
for x in QuestionInfo:
    print(x[1])
```
